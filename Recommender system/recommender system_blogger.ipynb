```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./recommender-2021-07-15.csv')
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df.rename(columns={'Unnamed: 0':'blogger_name'},inplace=True)
```

```python
df.index = df['blogger_name']
```

```python
df.drop('blogger_name',axis=1,inplace=True)
```

```python
# 글이 0개인 인플루언서는 지운다.
bool_df = df.apply(lambda x: x.sum() != 0 ,axis=1)
df = df[bool_df]
df
```

```python
influencer_based_collabor = cosine_similarity(df)
```

```python
influencer_based_collabor
```

```python
influencer_based_collabor = pd.DataFrame(data = influencer_based_collabor, index=df.index,columns=df.index)
```

```python
influencer_based_collabor
```

```python
def get_influencer_based_collabor(_blogger_name):
    return influencer_based_collabor[_blogger_name].sort_values(ascending=False)
get_influencer_based_collabor('clover_sky')
```
